```python
productos = {}

while True:
    print("\n1. Agregar producto")
    print("2. Mostrar productos")
    print("3. Buscar producto")
    print("4. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Nombre del producto: ")
        precio = float(input("Precio: "))
        productos[nombre] = precio

    elif opcion == "2":
        for nombre, precio in productos.items():
            print(nombre, ":", precio)

    elif opcion == "3":
        buscar = input("Producto a buscar: ")
        if buscar in productos:
            print("Precio:", productos[buscar])
        else:
            print("No encontrado")

    elif opcion == "4":
        break
```

PREGUNTAS DE REFLEXIÓN
1. ¿Qué diferencia hay entre lista y tupla?

Lista: se puede modificar (agregar, eliminar, cambiar elementos).

Tupla: es inmutable (no se puede cambiar después de creada).

2. ¿Cuándo es mejor usar diccionarios?

Cuando necesitas:

Asociar clave → valor

Buscar datos rápidamente

Representar información tipo “registro” (como estudiantes, productos, etc.)

3. ¿Qué estructura te pareció más útil?

Depende del problema:

Listas → para colecciones simples

Tuplas → datos fijos

Diccionarios → datos organizados y con significado

```python
lista = []

while True:
    print("\n1. Agregar elemento")
    print("2. Mostrar lista")
    print("3. Eliminar elemento")
    print("4. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        elemento = input("Elemento a agregar: ")
        lista.append(elemento)

    elif opcion == "2":
        print(lista)

    elif opcion == "3":
        eliminar = input("Elemento a eliminar: ")
        if eliminar in lista:
            lista.remove(eliminar)
        else:
            print("No está en la lista")

    elif opcion == "4":
        break
```
