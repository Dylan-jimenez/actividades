```python
cadena_principal = input("Introduce la cadena principal: ")
subcadena = input("Introduce la subcadena para comprobar: ")

if cadena_principal.startswith(subcadena):
    print(f"La cadena '{cadena_principal}' comienza con la subcadena '{subcadena}'.")
else:
    print(f"La cadena '{cadena_principal}' NO comienza con la subcadena '{subcadena}'.")
```

```python
cadena = input("Introduce la cadena principal: ")

caracter = ""
while len(caracter) != 1:
    caracter = input("Introduce un solo caracter para buscar: ")
    if len(caracter) != 1:
        print("Por favor, introduce *un solo* caracter.")

conteo = cadena.count(caracter)
print(f"El caracter '{caracter}' aparece {conteo} veces en la cadena '{cadena}'.")
```

```python
frase = input("Introduce una frase: ")

palabras = frase.split() # Divide la frase en una lista de palabras usando los espacios como delimitadores

conteo_palabras = len(palabras)

print(f"La frase tiene {conteo_palabras} palabras.")
```

```python
cadena = input("Introduce la cadena principal: ")

caracter1 = ""
while len(caracter1) != 1:
    caracter1 = input("Introduce el primer caracter (el que quieres reemplazar): ")
    if len(caracter1) != 1:
        print("Por favor, introduce *un solo* caracter para el primero.")

caracter2 = ""
while len(caracter2) != 1:
    caracter2 = input("Introduce el segundo caracter (por el que quieres reemplazar): ")
    if len(caracter2) != 1:
        print("Por favor, introduce *un solo* caracter para el segundo.")

cadena_modificada = cadena.replace(caracter1, caracter2)

print(f"La cadena original era: '{cadena}'")
print(f"El caracter '{caracter1}' ha sido reemplazado por '{caracter2}'.")
print(f"La cadena modificada es: '{cadena_modificada}'")
```
